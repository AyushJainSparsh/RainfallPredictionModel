```python
# As in preprocessing we get that annual rainfall is highly collinear with July August September and October rainfall data.
# So we prepare a model in which we give annual Rainfall and it will predict how much we get rainfall in Monsoon Months.
# As it is Regression problem so we use linear Regression firsts then we modify it by hypertuning and try to get a best model.
```

# importing everything

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
dataset = pd.read_csv(r'WestUttarPradeshClimate.csv')
```

```python
dataset.head()
```

# Feature Selection

```python
# In western part monsoon seasons are july august september and october
```

```python
dataset = dataset[['ANNUAL','JUL','AUG','SEP','OCT']]
```

```python
dataset
```

```python
sns.regplot(x = 'ANNUAL' , y = 'JUL' , data = dataset)
plt.title('Regresion plot between annual rainfall and july month')
#plt.savefig("AnnualVsJuly.png")
plt.show()
```

```python
sns.regplot(x = 'ANNUAL' , y = 'AUG' , data = dataset)
plt.title('Regresion plot between annual rainfall and august month')
#plt.savefig("AnnualVsAugust.png")
plt.show()
```

```python
sns.regplot(x = 'ANNUAL' , y = 'SEP' , data = dataset)
plt.title('Regresion plot between annual rainfall and september month')
#plt.savefig("AnnualVsSeptember.png")
plt.show()
```

```python
sns.regplot(x = 'ANNUAL' , y = 'OCT' , data = dataset)
plt.title('Regresion plot between annual rainfall and october month')
#plt.savefig("AnnualVsOctober.png")
plt.show()
```

```python
X = dataset[['ANNUAL']]
y = dataset[['JUL','AUG','SEP','OCT']]
```

# use train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=42)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
model = LinearRegression()
model.fit(X_train , y_train)
pred = model.predict(X_test)
print('mean_squared_error = ',mean_squared_error(y_test , pred))
print('r2_score = ',r2_score(y_test , pred))
print('mean_absolute_error = ',mean_absolute_error(y_test , pred))
```

```python
model.score(X_test , y_test)
```

```python
plt.scatter(y_test , pred , color = 'g')
plt.show()
```

# Multi Task Elastic Net CV

```python
from sklearn.linear_model import MultiTaskElasticNetCV
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
Selected_Model = MultiTaskElasticNetCV(cv = 5)
Selected_Model.fit(X_test , y_test)
pred = Selected_Model.predict(X_test)
print('mean_squared_error = ',mean_squared_error(y_test , pred))
print('r2_score = ',r2_score(y_test , pred))
print('mean_absolute_error = ',mean_absolute_error(y_test , pred))
```

# MultiTaskElasticNetCV with Grid Search

```python
from sklearn.linear_model import MultiTaskElasticNetCV
from sklearn.model_selection import GridSearchCV
model = MultiTaskElasticNetCV()

l1_ratio = [.5, .7, .9, .95, .99, 1]
cv = [2,10,5]
selection = ['cyclic', 'random']
params = dict(l1_ratio = l1_ratio , selection = selection)

grid = GridSearchCV(estimator=model, param_grid=params, scoring='r2')
```

```python
grid
```

```python
grid.fit(X_train , y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
pred = grid.predict(X_test)
print('mean_squared_error = ',mean_squared_error(y_test , pred))
print('r2_score = ',r2_score(y_test , pred))
print('mean_absolute_error = ',mean_absolute_error(y_test , pred))
```
